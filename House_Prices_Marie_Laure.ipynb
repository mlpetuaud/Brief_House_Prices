# Régression linéaire univariée

## Entrainement sur la totalité du dataset

### Avec StatsModel

```python
import statsmodels.formula.api as smf

# Fit regression model (using the natural log of one of the regressors)
results = smf.ols('SalePrice ~ OverallQual', data=data).fit()

# Inspect the results
print(results.summary())
```

**Interprétation des résultats**
- Le R2 permet de dire que, simplement avec la qualité d'ensemble des biens, nous sommes en mesure d'expliquer 62,5% de la variabilité des prix de vente autour de la moyenne. La p-value associée à la F-stat est très proche de 0, ce qui signifie que notre R2 est significatif.
- L'intercept et le coefficient associés à la qualité d'ensemble sont très faibles, ce qui est lié à la différence d'échelle entre la qualité (qui va de 1 à 10) et le prix des maisons (qui est en moyenne de 180K dollars). En revanche ils sont significatifs (p-value associée quasi nulle)
- ==> Compte tenu de ca, et du fait que la valeur des maisons présente des valeurs extrêmes sur les prix les plus élevés, se pose la question du passage de la variable à expliquer en logarithme. Regardons maintenant ce que ça donne :


```python
#testons donc l'impact sur la distribution d'un passage en log
data["Log_SalePrice"] = np.log(data["SalePrice"]) # ERROR ("A value is trying to be set on a copy of a slice from a DataFrame".)
sns.displot(data=data, x="Log_SalePrice").set(title="Distribution ");
```

```python
import statsmodels.formula.api as smf

# Fit regression model (using the natural log of one of the regressors)
results_log = smf.ols('Log_SalePrice ~ OverallQual', data=data).fit()

# Inspect the results
print(results_log.summary())
```

**Le passage en log a permis de d'accroitre de 4 points le R2**. Nous décidons donc de travailler dorénavant sur la prédiction des Log_SalePrice.
- Le **R2 de 0.668** signifie que 66,8% de la variabilité des prix de vente des maisons autour de leur moyenne est expliquée par notre modèle. Le reste est constitué des erreurs, considérées comme étant du bruit (à valider avec test de normalité des erreurs)
- La **p-value associée à la F-stat** permet de valider la significativité du R2.
- L'**Intercept de 10,5** correspond au log du prix de vente constaté lorsque la qualité est nulle. 
- Le **coefficient de 0,236** signifie que, pour une augmentation de la qualité de 1, le log du prix des maisons s'accroit de 0,236.  
- La **p-value associée à la t-stat** estime la **significativité de notre feature**. En l'occurence  elle est quasi nulle, ce qui signifie qu'on peut rejeter l'hypothèse nulle d'absence de significativité de notre coefficient.

<br/>Regardons **l'impact sur la distribution des résidus :**

```python
# Calcul des résidus
y_hat = results.predict(data["OverallQual"]) # on calcule les y prédits 
residus = y_hat - data["SalePrice"] # les résidus : différence entre la valeur déterminée par le modèle et la vraie valeur

# Distribution des résidus (histogramme)
#fig.title("Distribution des résidus - SalePrice en fonction de OverallQual")
sns.displot(data=residus, kde=True).set(title="Distribution des résidus - SANS Log");
```

```python
# Calcul des résidus
y_hat_log = results_log.predict(data["OverallQual"]) # on calcule les y prédits 
residus_log = y_hat_log - data["Log_SalePrice"] # les résidus : différence entre la valeur déterminée par le modèle et la vraie valeur

# Distribution des résidus (histogramme)
sns.displot(data=residus_log, kde=True).set(title="Distribution des résidus - AVEC Log");
```

```python
# Distribution des résidus (qqplot)
import pylab 
import scipy.stats as stats

stats.probplot(residus_log, dist="norm", plot=pylab)
pylab.show()
```

Le passage en log a également un impact positif sur la distribution des résidus. Comme on peut le voir ci-dessous, la variance et la moyenne des résidus semble également stable au regard de l'évolution de y_hat

```python
# hétéroscédasticité des résidus
residus_y = pd.DataFrame({'residus':residus_log, 'y':y_hat_log})
sns.scatterplot(data=residus_y, x=residus_y.y, y=residus_y.residus)
```

## Sélection d'un test et d'un train

Séparons maintenant notre dataset en 2 :
- un train_set qui prendra 80% des données et sur lequel le modèle s'entrainera. Ce sera sur la base du train que les paramètres du modèle (intercept et coefficients) seront déterminés.
- un test_set qui prendra 20% des données et qui servira à tester la qualité de notre modèle. On appliquera ensuite les paramètres appris sur le train aux Xtest pour mesurer à nouveau le R2.
- ==> En toute logique le R2 du test devrait être inférieur à celui du train. En effet le modèle est censé avoir de moins bons résultats en prédisant sur la base d'informations qu'il n'a jamais vues. L'objectif est que cet écart ne soit pas trop grand (sinon ca signifierait qu'il a appris par coeur sans être en mesure de généraliser)

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(data, test_size=0.2, random_state=42)
x_train = train_set[["OverallQual"]]
y_train = train_set["Log_SalePrice"]
x_test = test_set[["OverallQual"]]
y_test = test_set['Log_SalePrice']
```

```python
from sklearn.model_selection import train_test_split
x = data["OverallQual"]
y = data['Log_SalePrice']
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size = 0.2, random_state  = 32)
```

```python
reg2 = linear_model.LinearRegression()
reg2.fit(x_train, y_train);
```

```python
# première méthode pour calculer le R2 sur les données du test_set :
y_hat = reg2.predict(x_test);
from sklearn.metrics import r2_score , mean_squared_error
score = r2_score(y_test , y_hat)
print("R2 Score : {}".format(score))
```

```python
# seconde méthode :
reg2.score(x_test,y_test)
```

```python
reg2.score(x_train,y_train)
```

On a un R2 qui est plus important sur l'échantillon de test que de train.

Interprétation des résultats :

- Les p-values sont toutes nulles, ce qui permet de conclure que nos coefficients sont significatifs.
- Les prix de vente sont d'autant plus élevés que qualité et superficie le sont.
- A CREUSER : multicolinéarité des deux features (on a en effet un taux de corrélation de 0.59 entre ces deux variables, mais en connaissant la signification des données il ne me semble pas que l'information de l'une soit comprise dans l'information de l'autre.
```python
# Calcul des résidus
y_hat_log = results_log.predict(data[["OverallQual","GrLivArea"]]) # on calcule les y prédits 
residus_log = y_hat_log - data["Log_SalePrice"] # les résidus : différence entre la valeur déterminée par le modèle et la vraie valeur
# Distribution des résidus (histogramme)
print(sns.displot(data=residus_log, kde=True))
```
